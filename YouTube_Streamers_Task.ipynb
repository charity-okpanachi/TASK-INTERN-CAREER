## Trend Analysis

```python
#Total Categories

df.Categories.value_counts()
```

```python
# Print mean of each category

grouped_df = df.groupby('Categories')  
grouped_df.mean().sort_values(by='Categories')

```

```python
# Subscribers for each category

df_categories = df.groupby('Categories')['Suscribers'].sum().sort_values(ascending=False).plot(kind='bar',figsize=(10,5),width=0.92)
plt.show()
df_categories
```

```python
# Top 10 Categories by Subscribers

df['Categories'].value_counts().head(10)
```

```python
most_popular_categories = df['Categories'].value_counts().head(10)
most_popular_categories.plot(kind = 'bar')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Top 10 Categories by Subscribers')
plt.show()
```

#### The most popular category by subscribers is Música y baile

```python
# Checking correlation between Suscribers, likes and comments

print(df['Suscribers'].corr(df['Likes']))
 
print(df['Likes'].corr(df['Comments']))
 
print(df['Suscribers'].corr(df['Comments']))
```

#### There is a weak positive correlation between the number of Subcribers and the number of Likes
#### Correlation between the number of Likes and Comments is moderate
#### There is barely any correlation between the number of subcribers and the number of comments 

## Audience study

#### tseries is the top streamer by subscribers, followed by MrBeast and CoComelon

## Content category

```python
#Calculate the number of streamers in each content category:

grouped_categories = df.groupby('Categories')
#print(grouped_categories.size())
grouped_categories.size()
```

```python
# Number of streamers in each content category in descending order
category_counts = df['Categories'].value_counts().to_frame('Count').sort_values('Count', ascending=False)
print(category_counts)
```

```python
# Visualizing the number of streamers in each content category
category_counts = df['Categories'].value_counts().sort_values(ascending=False)

# Create a bar chart with different colors
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color=plt.cm.tab20(range(len(category_counts))))
plt.title('Number of Streamers in Each Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

#### Música y baile has the highest number of streamers but there are other categories which performed well such as                         
#### Películas, Animación                   
#### Música y baile, Películas              
#### Vlogs diarios                          
#### Noticias y Política

## Benchmarking

```python
# Streamers with above average performance in terms of subscribers

above_avg_df = df[df['Suscribers'] > df['Suscribers'].mean()]
above_avg_df
```

```python
# Streamers with above average performance in terms of visits

above_avg_df2 = df[df['Visits'] > df['Visits'].mean()]
above_avg_df2
```

```python
# Streamers with above average performance in terms of likes

above_avg_df3 = df[df['Likes'] > df['Likes'].mean()]
above_avg_df3
```

```python
# Streamers with above average performance in terms of comments

above_avg_df4 = df[df['Comments'] > df['Comments'].mean()]
above_avg_df4
```

#### In terms of subscribers, 203 streamers performed above average with 'tseries' taking the lead
#### 134 streamers in terms of visits. 'MrBeast' is topping the list
#### 119 streamers in terms of likes. 'MrBeast' still topping the list
#### 112 streamers in terms of comments. 'MrBeast' still leading the list
